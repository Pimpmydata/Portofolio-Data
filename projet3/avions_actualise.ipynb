```python
import pandas as pd
import ast
import datetime
import subprocess
import sys
import traceback
from datetime import timedelta
```

```python
#dataset original "composants" (en locale)

df_avions = pd.read_csv(r"C:\Users\bapti\Desktop\WCS Python\Projet 3\Projet-3-Skyanalytics\donnees_depart\aeronefs_2024-06-02.csv")
```

```python
#dataset actualisé de "logs_vols". Ça prend toujours la version publié le jour

base_url = 'http://sc-e.fr/docs/logs_vols_'
format_date = "%Y-%m-%d"
date_dep = datetime.datetime.strptime(datetime.datetime.now().strftime(format_date),format_date)
date_fin = datetime.datetime.strptime("2024-06-03", format_date)
data_avion_global = df_avions.copy()
df_global_global = pd.DataFrame()
while date_dep >= date_fin : 
    date_dep_str = date_dep.strftime(format_date)
    url = base_url + date_dep_str + '.csv'
    try:
        log_vols_today = pd.read_csv(url)
        data_avion_global = df_avions.merge(log_vols_today,right_on="aero_linked", left_on="ref_aero",how="left")
        df_global_global = pd.concat([data_avion_global, df_global_global], axis = 0)

    except Exception as e:
            print(e)
            print(url)
            traceback.print_exc()
    date_dep = date_dep - datetime.timedelta(days=1)


```

```python
df_global_global.dropna(subset="jour_vol",inplace=True)
df_global_global.sample(25)
```

```python
#merge entre le dataset "composants" et le dataset "degradation"


#drop colonnes avec information répété ou pas relevant. un peu de nettoyage aussi.

df_global_global.drop(columns=["aero_linked", "ref_vol"], inplace=True)
df_global_global["jour_vol"] = pd.to_datetime(df_global_global["jour_vol"])
df_global_global["end_maint"] = df_global_global["end_maint"].str.replace(" 00:00:00","")
df_global_global["last_maint"] = df_global_global["last_maint"].str.replace(" 00:00:00","")
df_global_global["last_maint"] = pd.to_datetime(df_global_global["last_maint"])
df_global_global["debut_service"] = pd.to_datetime(df_global_global["debut_service"])  
df_global_global["end_maint"] = pd.to_datetime(df_global_global["end_maint"])
```

```python
df_global_global[["end_maint","last_maint","en_maintenance"]][df_global_global["en_maintenance"]==True]
```

```python
df_global_global = df_global_global = df_global_global.dropna(subset=["etat_voyant"])
df_global_global["etat_voyant"]
```

```python
#change tous les avions qui ont un voyant allumé pour "en maitenance" et calcule la date de fin de maintenante
today = datetime.datetime.now().date()
df_global_global.loc[df_global_global["etat_voyant"] != 0, "en_maintenance"] = True
df_global_global.loc[df_global_global["etat_voyant"] == 1, "end_maint"] = pd.to_datetime(today + timedelta(days=1))
df_global_global.loc[df_global_global["etat_voyant"] == 2, "end_maint"] = pd.to_datetime(today + timedelta(days=7))
df_global_global.loc[df_global_global["etat_voyant"] == 3, "end_maint"] = pd.to_datetime(today + timedelta(days=14))
```

```python
df_global_global[df_global_global["etat_voyant"] != 0]
```

```python
#fonction pour calculer le cout de l'arret selon le voyant. Il faut encore ajouter le cout des composants

def cout(df):
    df["cout_avion"] = 0
    df.loc[df["etat_voyant"] == 3, "cout"] = 15000 * 14
    df.loc[df["etat_voyant"] == 2, "cout"] = 15000 * 7
    df.loc[df["etat_voyant"] == 1, "cout"] = 15000
    
    return df
    
```

```python
df_global_global = cout(df_global_global)
```
